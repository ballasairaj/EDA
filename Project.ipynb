# Tasks to Perform

## 1.  Handle Missing Data

- Identify which columns have missing or null values

```python
ds.isnull().sum()
```

- My data set consists of Null value columns (County, City, Postal Code, Electric Range, Base MSRP, Legislative District, Vehicle Location, Electric Utility, 2020 Census Tract)

### Choose the appropriate strategy:


- Remove rows or columns with too many missing entries.
   
- Fill missing values using suitable methods such as:
  
- Mean or median (for numerical values)

```python
ds["Base MSRP"].mean()
```

```python
ds["Base MSRP"] = ds["Base MSRP"].fillna(round(ds["Base MSRP"].mean()))
```

```python
ds["Electric Range"] = ds["Electric Range"].fillna(round(ds["Electric Range"].mean()))
```

```python
ds["Legislative District"] = ds["Legislative District"].fillna(round(ds["Electric Range"].mean()))
```

```python
ds["Postal Code"] = ds["Postal Code"].fillna(round(ds["Postal Code"].mean()))
```

- Filling the Numerical null value columns with the mean value

- Mode or placeholder (for categorical values)

```python
ds["County"] = ds["County"].fillna("Na")
```

```python
ds["City"] = ds["City"].fillna("NA")
```

```python
ds["Vehicle Location"] = ds["Vehicle Location"].fillna("unknown")
```

```python

ds["Electric Utility"] = ds["Electric Utility"].fillna("Unknown")
```

```python

ds["2020 Census Tract"] = ds["2020 Census Tract"].fillna("Unknown")
```

```python
ds.isnull().sum()
```

- Handling the missing data is done 
- There ate no null value columns present in my data set

# Multivariate 

### 1. Scatter Plot with Hue & Size

- Relationship between Model Year, Electric Range, and Base MSRP, colored by EV Type.

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=ds, 
                x="Model Year", 
                y="Electric Range", 
                size="Base MSRP", 
                hue="Electric Vehicle Type",
                alpha=0.6, palette="rainbow")
plt.title("Model Year vs Electric Range vs Base MSRP by EV Type")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show();
```


### Heatmap (Correlation Matrix for Numerical Features)

- Shows relationships among Model Year, Electric Range, Base MSRP, Census Tract, Legislative District.


```python
plt.figure(figsize=(8,6))
corr = ds[["Model Year","Electric Range","Base MSRP","Legislative District","2020 Census Tract"]].corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap of Numerical Variables")
plt.show();
```

###  Stacked Bar Plot (Categorical  Categorical)

-  Distribution of Make across EV Types.

```python
cross_tab = pd.crosstab(ds["Make"], ds["Electric Vehicle Type"])
cross_tab.head(10).plot(kind="bar", stacked=True, figsize=(10,6), colormap="rainbow")
plt.title("Top 10 Makes by EV Type (Stacked Bar)")
plt.xticks(rotation=45)
plt.show();
```

###  2. Line Plot (Trends over Time)

 - Trend of Average Electric Range per Model Year, separated by EV Type.

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=ds, 
             x="Model Year", 
             y="Electric Range", 
             hue="Electric Vehicle Type", 
             estimator="mean", 
             palette="rainbow", marker="o")
plt.title("Average Electric Range by Model Year and EV Type")
plt.show();
```

### Bar Plot (Categorical  Categorical)

-  Compare Top 10 Makes by EV Type using a grouped bar plot.

```python
top_makes = ds["Make"].value_counts().index[:10]
plt.figure(figsize=(10,5))
sns.countplot(data=ds[ds["Make"].isin(top_makes)], 
              x="Make", 
              hue="Electric Vehicle Type", 
              palette="rainbow")
plt.title("Top 10 Vehicle Makes by EV Type")
plt.xticks(rotation=45)
plt.show();
```

## Strip Plot  Base MSRP by EV Type

- Compare the price (Base MSRP) distribution across EV Types

```python
plt.figure(figsize=(10,6))
sns.stripplot(data=ds, 
              x="Electric Vehicle Type", 
              y="Base MSRP",
              jitter=True,      # spreads points for readability
              alpha=0.5,        # transparency for overlap
              color="blue")

# Set y-axis limits
plt.ylim(0, 200000)

# Add y-axis ticks every 50,000
plt.yticks(range(0, 200001, 50000))

plt.title("Strip Plot of Base MSRP by EV Type")
plt.show();
```
